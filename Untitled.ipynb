```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('Output_Images_intervals/frame_10020_object_1.jpeg')
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
sift = cv2.SIFT_create()
original_keypoints, original_descriptor = sift.detectAndCompute(gray_image, None)
keypoints_without_size = np.copy(image)
keypoints_with_size = np.copy(image)

img1 = cv2.drawKeypoints(image, original_keypoints, keypoints_without_size, color = (0, 255, 0))
img2 = cv2.drawKeypoints(image, original_keypoints, keypoints_with_size, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
fast = cv2.FastFeatureDetector_create() 
keypoints_with_nonmax = fast.detect(gray_image, None)
fast.setNonmaxSuppression(False)
keypoints_without_nonmax = fast.detect(gray_image, None)
image_with_nonmax = np.copy(image)
image_without_nonmax = np.copy(image)
img3 = cv2.drawKeypoints(image, keypoints_with_nonmax, image_with_nonmax, color=(0,255,0),
flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
img4 = cv2.drawKeypoints(image, keypoints_without_nonmax, image_without_nonmax, color=(0,255,0), 
flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.imshow(img3)
```

```python
plt.imshow(img4)
```

```python
orb = cv2.ORB_create()
original_keypoints, original_descriptor = orb.detectAndCompute(gray_image, None)
keypoints_without_size = np.copy(image)
keypoints_with_size = np.copy(image)

img5 = cv2.drawKeypoints(image, original_keypoints, keypoints_without_size, color = (0, 255, 0))
img6 = cv2.drawKeypoints(image, original_keypoints, keypoints_with_size, flags = 
cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.imshow(img5)
```

```python
plt.imshow(img6)
```
